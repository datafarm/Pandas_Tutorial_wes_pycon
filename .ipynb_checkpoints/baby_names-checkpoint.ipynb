```python
def get_diversity(group):
    df3 = group.sort_index(by='prop',ascending=False)
    return df3.prop.cumsum().searchsorted(0.5)
boys.groupby('year').apply(get_diversity)
```

```python
boys.groupby('year').apply(get_diversity).plot()
```

```python
boys_count = boys.groupby('year').apply(get_diversity)
girls_count = girls.groupby('year').apply(get_diversity)
```

```python
boys_count.plot(label='boy')
girls_count.plot(label='girl')
legend(loc='best')
```
